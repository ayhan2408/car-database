```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/car-data/cardata.csv")
car = df.copy()
car.head(10)
```

```python
car["Car_Name"].value_counts()
#burda araba markalarına gore kac tane var onun degerlerine bakıldı.
```

```python
select_car = "city"
car_city = car[car["Car_Name"] == select_car]
car_city.to_csv("car_city.csv"  , index = False)
car_city.head()
```

```python
car_city.value_counts
```

```python
sns.barplot(x = "Car_Name" , y = "Present_Price" , hue = "Transmission" ,   data = car_city)
```

```python
sns.catplot(x = "Year" , y = "Kms_Driven" , data = car_city);
```

```python
sns.scatterplot(x = "Kms_Driven" , y = "Present_Price" , hue = "Fuel_Type" , data = car_city)
```

```python
sns.boxplot(x = "Fuel_Type" , y = "Selling_Price" , hue = "Year" , data = car)
plt.xticks(rotation = 90);
```

```python
sns.scatterplot(x = "Selling_Price" , y = "Present_Price" , hue = "Year" , data = car);
plt.xticks(rotation = 90);
```

```python
select_year = 2016
car_2016 = car[car["Year"] == select_year]
car_2016.to_csv("car_2016.csv" , index = True)
car_2016
```

```python
car_2016["Fuel_Type"].value_counts()
```

```python
car_2016["Car_Name"].value_counts()
```

```python
car_2016_Petrol["Car_Name"].value_counts()
```

```python
sns.barplot(x = "Car_Name" , y = "Present_Price" , hue = "Transmission" , data = car_2016_Petrol);
plt.xticks(rotation = 90);
#burda 2016 Benzinli olan aracların satıs fiyatları uzerinden degerlendirmler yaptık .
```
